## Use on the prompt classification dataset

- I had used this dataset previously but noticed that it's quite bad, has some duplicates (and wrong answers), so for curiosity see if semhash captures this

```python
from datasets import load_dataset
from semhash import SemHash

# Load a dataset to deduplicate
texts = load_dataset("data-is-better-together/10k_prompts_ranked", split="train")["prompt"]

# Initialize a SemHash instance
semhash = SemHash.from_records(records=texts)

# Deduplicate the texts
deduplicated_texts = semhash.self_deduplicate().deduplicated
```
